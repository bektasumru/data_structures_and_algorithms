A Binary Search Tree (BST) is a fundamental data structure in computer science designed to store and manage data in a sorted, hierarchical format. In a BST, each node has up to two child nodes: a left and a right. The left child holds values that are less than the parent node, while the right child contains values that are greater than the parent.

This ordering rule makes it possible to perform operations like searching, insertion, and deletion with high efficiency—typically in logarithmic time, assuming the tree is balanced. Thanks to this structure, BSTs are widely used in applications where quick lookups and sorted data access are important, such as databases, file systems, and real-time systems.

Binary Search Tree Project
Given the array:
[7, 5, 1, 8, 3, 6, 0, 9, 4, 2]

We construct a Binary Search Tree (BST) by inserting each value one by one, following the BST rule:

If the value is less than the current node, go to the left; if greater, go to the right.

Explanation:

7 becomes the root node.

5 is less than 7 → it goes to the left of 7.

1 is less than 5 → it goes to the left of 5.

8 is greater than 7 → it goes to the right of 7.

3 is greater than 1 → it goes to the right of 1.

6 is greater than 5 → it goes to the right of 5.

0 is less than 1 → it goes to the left of 1.

9 is greater than 8 → it goes to the right of 8.

4 is greater than 3 → it goes to the right of 3.

2 is less than 3 → it goes to the left of 3.
